```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
value_20 = np.loadtxt("N20_p20_R5_T5.txt")
value_10 = np.loadtxt("N10_p10_R5_T5.txt")
# value_5 = np.loadtxt("N20_p5_R5.txt")
# value_1 = np.loadtxt("N2_p1_R5.txt")
```

```python
### This creates the plots

fig = plt.figure(figsize=(10, 6), dpi=200)
ax = fig.add_subplot(111)


plt.semilogx(value_20, label='N = 20')
plt.semilogx(value_10, label='N = 10')
# plt.semilogx(value_5, label='N = 20')
plt.xlabel('Time')
plt.ylabel('$<R_p(t)R_p(0)>/<|R_p(0)|^2>$')
plt.legend(loc='best')
plt.show()
```
